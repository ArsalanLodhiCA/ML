```python
import pandas as pd

# Read a CSV file into a DataFrame
df = pd.read_csv('./csv/1098107969.csv')
fewcols_df = df.drop(columns=['Unnamed: 0','Reviewer', 'ReviewerURL', 'VerifiedPurchase','HelpfulCount'])

print(fewcols_df.dtypes)
fewcols_df.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

# Create a word cloud from the column 'text' in the DataFrame
reviews_text = " ".join(review for review in fewcols_df.Description)
stopwords = set(stopwords.words("english"))
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(reviews_text)


# Display the word cloud
plt.figure(figsize=[20,10])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
